# feature 추가

```python
df_copy = df.copy()
snow_ct = [-0.1, 4.0, 100.0]
snow_ct_name = [0, 1]
df_copy['snow_ct'] = pd.cut(df_copy['snow'], bins=snow_ct, labels=snow_ct_name)
df_copy
```

```python
df_copy['snow_ct'].value_counts()
```

```python
df_drop = df_copy.drop(['wind_direction', 'local_pressure', 'vapor_pressure', 'insolation'], axis=1)
```

```python
def data_prep(df, valid=False):
    if valid==False:
        df_train = df.loc[:26303]
        df_test = df.loc[26304:]
        
        X_train = df_train.drop(['date', 'generation'], axis=1)
        y_train = df_train['generation']
        
        X_test = df_test.drop(['date', 'generation'], axis=1)
        y_test = df_test['generation']
        
        return X_train, y_train, X_test, y_test, df_test
    
    else:
        df_train = df.loc[:17519]
        df_valid = df.loc[17520:26303]
        df_test = df.loc[26304:]
        
        X_train = df_train.drop(['date', 'generation'], axis=1)
        y_train = df_train['generation']
        
        X_valid = df_valid.drop(['date', 'generation'], axis=1)
        y_valid = df_valid['generation']
        
        X_test = df_test.drop(['date', 'generation'], axis=1)
        y_test = df_test['generation']
        
        return X_train, y_train, X_valid, y_valid, X_test, y_test, df_valid, df_test
```

```python
X_train, y_train, X_test, y_test, df_test = data_prep(df_drop)
```

```python
X_train
```

```python
def train_model(model, df, X_train, y_train, X_test, y_test):
    model = model
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    df_pred = pd.DataFrame({'date':df['date'],
                        'time':df['time'],
                        'test':y_test,
                        'pred':y_pred})
    df_pred['error'] = df_pred['test'] - df_pred['pred']
    
    scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
    rmse = np.sqrt(-scores).mean()    
    
    return df_pred, rmse
```

```python
model = RandomForestRegressor()

df_pred, rmse = train_model(model, df_test, X_train, y_train, X_test, y_test)
```
