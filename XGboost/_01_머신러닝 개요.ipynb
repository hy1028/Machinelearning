### 1.2.3 누락된 값 처리하기

#### 누락된 값의 개수 계산하기

```python
data.isnull().sum().sum() # 누락된 값의 전체 개수
```

```python
data[data.isnull().any(axis=1)] # 누락된 값이 들어 있는 모든 행 출력
```

#### 누락된 값 고치기

```python
# 중간값이나 평균으로 바꾸기
data['windspeed'].fillna(data['windspeed'].median(), inplace=True)
```

```python
data.iloc[[56, 81]] # windspeed의 결측치가 median값으로 채워진 것을 알 수 있다. 
```

```python
data.groupby(['season']).median() # season별로 그룹을 묶고 중간값으로 집계하였다. 
```

```python
data['hum'] = data['hum'].fillna(data.groupby('season')['hum'].transform('median')) # hum칼럼의 결측치를 season별 중간값으로 채움
```

```python
data.iloc[[129, 213, 388]] # hum의 결측치가 채워진 것을 확인
```

```python
data[data['temp'].isnull()]
```

```python
# temp와 atemp의 누락값을 어제와 내일 온도의 평균으로 채운다. 
mean_temp = (data.iloc[700]['temp'] + data.iloc[702]['temp'])/2
mean_atemp = (data.iloc[700]['atemp'] + data.iloc[702]['atemp'])/2

data['temp'].fillna(mean_temp, inplace=True)
data['atemp'].fillna(mean_atemp, inplace=True)
```

```python
data.iloc[[701]] # 결측치 채움 확인
```

```python
# 날짜 객체를 추정하려면 datatime type을 사용해야하나 dteday의 type은 object로 확인됨.
# to_datetime() methode를 사용해 type 변경 가능.

data['dteday'] = data['dteday'].apply(pd.to_datetime, infer_datetime_format=True, errors='coerce')
# infer_datetime_format=True로 지정하면 판다스가 datetime 객체의 종류를 결정하며 대부분의 경우 안전함.
```

```python
import datetime as dt
```

```python
data['mnth'] = data['dteday'].dt.month # data['mnth'] 값을 month값으로 일괄 재정리하였다. 
```

```python
data.tail()
```

```python
data.loc[730, 'yr'] = 1.0 # yr 칼럼은 이미 정규화되어있기 때문에 2012의 정규화값인 1로 같이 채워넣었다. 
```

## 1.3 회귀 모델 만들기

```python
data = data.drop(['casual', 'registered'], axis=1) # 정답값인 'cnt'는 casual과 registered의 합으로 이루어져 있기 때문에 삭제함
```

```python
data.to_csv('bike_rentals_cleaned.csv', index=False) # 데이터 랭글링 완료 된 데이터셋 저장
```

```python
X = data.drop(['cnt'], axis=1) # feature
y = data['cnt'] # target
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
import warnings
warnings.filterwarnings('ignore') # sklearn 버전 변경 사항 알림 경고 끄기
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test, y_pred) # 평균 제곱근 오차 구함, squared=False 넣으면 RMSE 바로 구할 수도 있음
rmse = np.sqrt(mse) # numpy로 제곱근 구함
```

```python
print(f'RMSE: {rmse:0.2f}')
```

```python
data['cnt'].describe() # RMSE와 비교하기 위해 정보 확인
```

오차가 작지 않은 것을 알 수 있음

### 1.3.8 XGBoost

```python
from xgboost import XGBRegressor
```

```python
xg_reg = XGBRegressor()
xg_reg.fit(X_train, y_train)
y_pred = xg_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'RMSE: {rmse:0.2f}')
```

##### 선형 회귀 교차 검증

```python
from sklearn.model_selection import cross_val_score

model = LinearRegression()
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
```

```python
rmse = np.sqrt(-scores)
```

```python
print('회귀 rmse:', np.round(rmse, 2))
print(f'RMSE 평균: {rmse.mean():0.2f}')
```

##### XGBoost 교차 검증

```python
model = XGBRegressor()
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
rmse = np.sqrt(-scores)
```

```python
print('회귀 rmse:', np.round(rmse, 2))
print(f'RMSE 평균: {rmse.mean():0.2f}')
```

### 1.4.3 데이터 랭글링 : 데이터셋 전처리

##### 데이터 적재

```python
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
```

```python
df_census # 인구 조사 소득 데이터셋으로 개인 소득 예측하기
```

```python
df_census.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hour-per-week', 
                  'native-country', 'income']
```

```python
df_census.head()
```

```python
df_census.info() # 누락값 확인(없음)
```

수치형이 아닌 열

```python
df_census = df_census.drop(['education'], axis=1) # education-num과 동일 정보라 삭제
```

```python
df_census = pd.get_dummies(df_census)
df_census.head()
```

```python
df_census = df_census.drop('income_ <=50K', axis=1)
```

```python
X = df_census.iloc[:, :-1]
y = df_census.iloc[:, -1]
```

```python
from sklearn.linear_model import LogisticRegression
```

교차 검증 함수

```python
def cross_val(classifier, num_splits=10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits) # sklearn 분류모델은 기본적으로 정확도로 모델을 평가하므로 scoring 매개변수 지정하지 않음
    print('정확도: ', np.round(scores, 2))
    print(f'평균 정확도: {scores.mean():0.2f}')
```

```python
cross_val(LogisticRegression())
```

```python
from xgboost import XGBClassifier

cross_val(XGBClassifier())
```
